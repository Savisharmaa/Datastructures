**Ques. What is causation? Explain difference between correlation and causation with an example.**

**Ans.** Causation (or causality) means that one variable directly affects another — in other words, a change in one variable causes a change in the other.

Correlation is a statistical relationship between two variables — they move together, but one does not necessarily cause the other.

Causation is a cause-effect relationship — a change in one variable directly causes a change in the other.

Example: Ice cream sales up.. drowning cases up = correlated but eating ice cream doesn't cause drowning = no causation.

A third variable is hot summer weather causes both:

More people buy ice cream and more people go swimming.

More people go swimming (leading to more drowning risks).
