**Ques. What is a random variable in probability theory?**

**Ans.** In probability theory, a random variable is a function that assigns a numerical value to each possible outcome of a random experiment.A random variable maps outcomes from a sample space (the set of all possible outcomes) to real numbers.

**Ques.What are the types of random variables?**

**Ans.** Types of Random Variables:

Discrete Random Variable: Takes on countable values (e.g., 0, 1, 2, ...).

Example: Number of heads in 3 coin tosses.

Continuous Random Variable: Takes on values from an interval of real numbers (uncountable).

Example: The time it takes for a light to turn green.

**Ques. What is the difference between discrete and continuous distributions?**

**Ans.** A discrete distribution deals with countable outcomes like number of students in a class.

A continuous distribution deals with uncountably infinite outcomes over an interval like height of people, time to complete a task.
**Ques. What are probability distribution functions (PDF)?**

**Ans.** A Probability Distribution Function (PDF) refers to a mathematical function that describes the likelihood of different outcomes for a random variable. The term PDF is most commonly used for continuous random variables, but there are analogous functions for discrete variables as well.

**Ques. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

**Ans.**PDF gives the probability of a random variable falling within range, while a CDF gives the probability that a random variable is less than or equal to a certain value.

**Ques. What is a discrete uniform distribution?**

**Ans.**A discrete uniform distribution is distribution where all outcomes have an equal chance of occuring. For example: rolling a dice, each number has same probability.
**Ques. What are the key properties of a Bernoulli distribution?**

**Ans.** The Bernoulli distribution has exactly two possible outcomes: success(1) and failure(2). It is defined by one parameter p which is the probability of success.

**Ques. What is the binomial distribution, and how is it used in probability?**

**Ans.** The binomial distribution is a fundamental probability distribution used to model the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes: success or failure.

**Ques. What is the Poisson distribution and where is it applied?**

**Ans.** The Poisson distribution is a discrete probability distribution that models the number of events occurring in a fixed interval of time or space, given that the events happen independently and at a constant average rate.

**Ques. What is a continuous uniform distribution?**

**Ans.**The continuous uniform distribution is a simple probability distribution used to model a situation where all outcomes within a certain range are equally likely. Example: A timer picks a random moment between 0 and 10 seconds

 **Ques.What are the characteristics of a normal distribution?**

**Ans.**A normal distribution is a bell shaped, symmetrical curve where most values are around the mean. It is defined by its mean and standard deviation. Many natural and social phenomena follow this pattern.

 **Ques.What is the standard normal distribution, and why is it important?**

**Ans.**A standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is important because it simplifies calculation and comparison in statistics using z scores.

 **Ques.What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

 **Ans.** The CLT states that the distribution of the sample mean will be approximately normal, regardless of the shape of the population, if the sample size is large enough. it is critical because it allows us to make predictions and conduct tests usiong normal distribution formulas.

 **Ques.How does the Central Limit Theorem relate to the normal distribution?**

 **Ans.** The CLT explains why the normal distribution appears so often in statistics, even when the original data is not normally distributed.

 **Ques.What is the application of Z statistics in hypothesis testing?**

**Ans.** The Z-statistic (or Z-score) measures how many standard deviations an observed sample statistic (usually a sample mean) is from the population parameter under the null hypothesis.

 **Ques.How do you calculate a Z-score, and what does it represent?**

**Ans.**A Z-score is calculated by subtracting the mean from a data point and dividing by the standard deviation. It shows how far and in what direction a data point is from the average in terms of standard deviation.
 **Ques.What are point estimates and interval estimates in statistics?**

 **Ans.**A point estimate is a single value used to estimate an unknown population parameter. It is calculated from sample data and serves as the "best guess" for that parameter. An interval estimate gives a range of values within which the population parameter is expected to lie, along with a confidence level.


 **Ques.What is the significance of confidence intervals in statistical analysis?**

 **Ans.** Confidence intervals provide a range of plausible values for a population parameter, reflecting uncertainty due to sampling variability.
 **Ques.What is the relationship between a Z-score and a confidence interval?**

 **Ans.**A z-score is used to calculate the limits of confidence interval. Higher Z-scores correspond to wider confidence intervals and greater confidence levels.
 **Ques.How are Z-scores used to compare different distributions?**

 **Ans.** z-scores convert different data points from different distributions into a common scale.This makes it easier to compare values from different datasets by seeing how far each value is from its own mean.

 3. Data should be randomly selected.

 **Ques.What is the concept of expected value in a probability distribution?**

 **Ans.** The expected value is the average value we expect to get if e repeat an experiment many times. It is calculated by multiplying each possible value by its probability and adding them together.

 **Ques.How does a probability distribution relate to the expected outcome of a random variable??**

 **Ans.**A probability distribution shows all possible outcomes and their probabilities. It helps to predict the expected outcome of a random variable by considering all outcomes and their chances.
